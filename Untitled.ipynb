# 2 Keypoints Using MP Holistic

```python
mp_holistic = mp.solutions.holistic #Holistic Model that is used to predict the Keypoints
mp_drawing = mp.solutions.drawing_utils #Drawing Utilities that will draw lines in the frame on the hand 
```

```python
# a function for detection only of the keypoints and its landmarks without showing lines
def mediapipe_detection(image,  model):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #to convert color space as for detection we need RGB while the cv give us BGR
    image.flags.writeable = False #Image is not writable or cannot be changeable
    results = model.process(image) #Detect the Holistic keypoints whre the image here is the frame from openCV
    image.flags.writeable = True #Image is now writable and can be changeable
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR) #to convert color space as for show in openCV webcam
    return image,results #as a result we will return the modified image and result back to the loop
```

```python
def draw_styled_landmarks(image,results):
    #Draw Face Connections
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_TESSELATION,
                             mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1), #Dot Colors
                             mp_drawing.DrawingSpec(color=(243,243,252), thickness=1, circle_radius=1) #Connections/Line Colors
                             ) 
    #Draw Pose Connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(10,150,12), thickness=1, circle_radius=1), #Dot Colors
                             mp_drawing.DrawingSpec(color=(213,213,212), thickness=1, circle_radius=1) #Connections/Line Colors
                             ) 
    #Draw Left Hand Connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(250,210,120), thickness=1, circle_radius=1), #Dot Colors
                             mp_drawing.DrawingSpec(color=(208,244,212), thickness=1, circle_radius=1) #Connections/Line Colors
                             )  
    #Draw Right Hand Connections
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(87,98,99), thickness=1, circle_radius=1), #Dot Colors
                             mp_drawing.DrawingSpec(color=(76,65,34), thickness=1, circle_radius=1) #Connections/Line Colors
                             )  
```

```python
cap = cv2.VideoCapture(0) #Accessing to the webcam we say it to capture video from device 0 

#Set mediapipe model
with mp_holistic.Holistic(min_detection_confidence=0.7, min_tracking_confidence =0.7) as holistic:
    while cap.isOpened(): #It is a confirmation that we access the Webcam and initiating a loop
        
        #Read Feed
        ret, frame = cap.read() #Reading a frame from the Webcam
    
        #Make Detection
        image , results = mediapipe_detection(frame,holistic)
        print(results)
        
        #Draw Styled Connections
        draw_styled_landmarks(image,results)
    
        #show to the Screen
        cv2.imshow('Webcam Frame',image)
        #Break the Loop Smoothly
        if cv2.waitKey(10) & 0xFF == ord('q'): #if q is pressed it will quit 
            break
    cap.release()
    cv2.destroyAllWindows()
```

# 3 Extract Keypoints Values

```python
def extract_keypoints(results):
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33*4)
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    left_hand = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    right_hand = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([pose, face, left_hand, right_hand])
```

```python
def extract_keypoints(results):
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33*4)
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([pose, face, lh, rh])
```

# 4 Setup Folder for Collection

```python
DATA_PATH = os.path.join('MP_Data') #Path for Exported Data, which will be in form of numpy arrays
actions = np.array(['Hello','I love You']) #Actions that we want to detect
no_sequences = 30 #30 videos or photos of data trained for each action
sequence_length = 30 #each of video will be of 30 frames
```

```python
for action in actions:
    for sequence in range(no_sequences):
        try:
            os.makedirs(os.path.join(DATA_PATH, action, str(sequence)))
        except:
            pass
```

# 5 Collect Keypoints Values for Training and Testing

```python
cap = cv2.VideoCapture(0) #Accessing to the webcam we say it to capture video from device 0 

#Set mediapipe model
with mp_holistic.Holistic(min_detection_confidence=0.7, min_tracking_confidence =0.7) as holistic:
    
    #Loop for actions
    for action in actions:
        #Loop for number of sequences/Video
        for sequence in range(no_sequences):
            #Loop for each frame in one sequence
            for frame_num in range(sequence_length):
                
                #Read Feed
                ret, frame = cap.read() #Reading a frame from the Webcam
    
                #Make Detection
                image , results = mediapipe_detection(frame,holistic)
                print(results)
        
                #Draw Styled Connections
                draw_styled_landmarks(image,results)
                
                #Apply break between each frame
                if frame_num == 0:
                    cv2.putText(image, 'Starting Collection', (120,200),
                                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 4, cv2.LINE_AA)
                    cv2.putText(image, 'Collecting Frames for {} Video number {}'.format(action,sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1, cv2.LINE_AA)
                    cv2.waitKey(2000)
                else:
                    cv2.putText(image, 'Collecting Frames for {} Video number {}'.format(action,sequence), (15,12),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1, cv2.LINE_AA)
                
                #Extract the Keypoints from frame and save to to a numpy array
                keypoints = extract_keypoints(results)
                npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_num))
                np.save(npy_path, keypoints)
                
                #show to the Screen
                cv2.imshow('Webcam Frame',image)
        
                #Break the Loop Smoothly
                if cv2.waitKey(10) & 0xFF == ord('q'): #if q is pressed it will quit 
                    break
    cap.release()
    cv2.destroyAllWindows()
```

# 6 Preprocess data and Create Labels and Features

```python
from sklearn.model_selection import train_test_split #partition of dataset into train dataset and test dataset
from tensorflow.keras.utils import to_categorical #to convert of catogory
```

```python
label_map = {label:num for num, label in enumerate(actions)}
```

```python
sequences, labels =[], [] #sequences is a X data feautures and labels as Y data or Class
for action in actions:
    for sequence in range(no_sequences):
        window = [] #all frames for a single sequence/video
        for frame_num in range(sequence_length):
            res = np.load(os.path.join(DATA_PATH, action, str(sequence), "{}.npy".format(frame_num)))
            window.append(res)
        sequences.append(window)
        labels.append(label_map[action])
```

```python
np.array(sequences).shape
```

```python
X = np.array(sequences)
```

```python
y = to_categorical(labels).astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05)
```

# 7 Build and Train LSTM Neural Network

```python
from tensorflow.keras.models import Sequential #a sequential neural network
from tensorflow.keras.layers import LSTM, Dense #a LSTM component to built a network and to made the prediction then pass to dense
from tensorflow.keras.callbacks import TensorBoard #perform or generate loggings to trace our model
```

```python
log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir = log_dir) #to monitor your Nural Network
```

```python
#To build Neural Network Architecture
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(30,1662)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(actions.shape[0], activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
model.fit(X_train, y_train, epochs=2000, callbacks=[tb_callback])
```
